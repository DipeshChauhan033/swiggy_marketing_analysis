```python
fake_na = ["", " ", "NA", "N/A", "na", "null", "None", "--"]

rd["Month"] = (
    rd["Month"]
    .astype(str)
    .str.strip()          # remove spaces
    .replace(fake_na, pd.NA)
)
```

```python
mode_month = rd["Month"].dropna().mode()[0]
rd["Month"] = rd["Month"].fillna(mode_month)
```

```python
rd.columns = rd.columns.str.lower()
```

```python
rd.columns
```

```python
rd = rd.rename(columns={'ctr':'click_through_rate',
                        'cpc':'click_per_rate',
                         'roi' : 'return_on_investment'})
```

```python
rd.columns
```

```python
#Create a New Column Performance_By_Month
```

```python
label = ['Good month','Average month','Bad month']
rd['performance_by_month'] = pd.qcut(rd['profit'],q = 3, labels = label)
```

```python
rd[['profit','performance_by_month']].head(10)
```

```python
#Removing Space from new column and convert into lower letters
```

```python
rd['performance_by_month'] = rd['performance_by_month'].str.replace(" ","_")
rd['performance_by_month'] = rd['performance_by_month'].str.lower()
```

```python
rd[['profit','performance_by_month']].head(20)
```

```python
#Total purchases
total_purchases = rd['purchases'].sum()
```

```python
print("Total_Purchases:",total_purchases)
```

```python
#Sum Of Revenue By Year & its Chart
rd["month"] = rd["month"].astype(str)
rd["year"] = rd["month"].str[:4]
rd.groupby("year")["revenue"].sum()
```

```python
rd.groupby("year")["revenue"].sum().plot()
plt.show()
```
